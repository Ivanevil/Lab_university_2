```python
# Гиперболическая аппроксимация

import numpy as np
import matplotlib.pyplot as plt

fid = open("./data_var07.txt", "r")
lines = fid.readlines()
fid.close()

y = [float(lines[i].split(',')[1]) for i in range(1, len(lines))]
x = [i for i in range(1, len(y)+1)]

n=len(y)
s=sum(y) # сумма значений y
s1=sum([1/y[i] for i in  range(0,n)]) #  сумма 1/x
s2=sum([(1/y[i])**2 for i in  range(0,n)]) #  сумма (1/x)**2
s3=sum([y[i]/y[i]  for i in range(0,n)])  # сумма y/x                   
a= round((s*s2-s1*s3)/(n*s2-s1**2),3) # коэфициент а с тремя дробными цифрами
b=round((n*s3-s1*s)/(n*s2-s1**2),3)# коэфициент b с тремя дробными цифрами
s4=[a+b/y[i] for i in range(0,n)] # список значений гиперболической функции              



plt.plot(x, y)
plt.plot(x, s4)


plt.grid(True)
plt.show()
```
